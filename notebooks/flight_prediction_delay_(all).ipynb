```python
import sys
from IPython import get_ipython

# Get a list of all defined variables in the current namespace
# Use 'user_vars' to filter out built-in IPython variables
variable_names = get_ipython().magic('%who_ls')

# Create a dictionary to store variable names and their sizes
variable_sizes = {}

for name in variable_names:
    # Use eval() to get the variable object and then sys.getsizeof()
    size_bytes = sys.getsizeof(eval(name))
    variable_sizes[name] = size_bytes

# Convert bytes to a more readable format (e.g., KB, MB)
def format_size(size_bytes):
    if size_bytes < 1024:
        return f"{size_bytes} B"
    elif size_bytes < 1024**2:
        return f"{size_bytes / 1024:.2f} KB"
    else:
        return f"{size_bytes / 1024**2:.2f} MB"

# Print the top 10 largest variables
print("--- Top 10 Largest Variables by Memory Usage ---")
sorted_vars = sorted(variable_sizes.items(), key=lambda item: item[1], reverse=True)

for name, size in sorted_vars[:10]:
    print(f"| {name:<20} | {format_size(size):>10} |")
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, dep,
    test_size=0.2,
    random_state=42
)

print(X_train.shape, X_test.shape)
```

```python
del(X)
```

```python
X_train.drop(columns=['Dep_DateTime', 'Arr_DateTime'], inplace=True)
X_test.drop(columns=['Dep_DateTime', 'Arr_DateTime'], inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import numpy as np

# -------------------------------------------
# 1. Identify categorical columns
# -------------------------------------------
cat_cols = X_train.select_dtypes(include=["object", "string"]).columns.tolist()
print("Categorical columns:", cat_cols)

# -------------------------------------------
# 2. Clean categorical missing values
# -------------------------------------------
def clean_cats(df):
    df = df.copy()
    for col in cat_cols:
        df[col] = df[col].astype("string")
        df[col] = df[col].fillna("missing")
        df[col] = df[col].replace({"<NA>": "missing", "nan": "missing", "None": "missing"})
    return df

X_train_clean = clean_cats(X_train)
X_test_clean  = clean_cats(X_test)

# -------------------------------------------
# 3. Apply LabelEncoder to categorical columns
#    (IMPORTANT: fit on train, transform on test)
# -------------------------------------------
encoders = {}  # store encoders so mapping stays consistent

for col in cat_cols:
    le = LabelEncoder()
    X_train_clean[col] = le.fit_transform(X_train_clean[col])

    # handle unseen categories in test
    X_test_clean[col] = X_test_clean[col].map(lambda x: x if x in le.classes_ else "missing")

    # extend encoder classes_ to include "missing" if needed
    if "missing" not in le.classes_:
        le.classes_ = np.append(le.classes_, "missing")

    X_test_clean[col] = le.transform(X_test_clean[col])
    encoders[col] = le

print("✅ Label encoding complete for all categorical variables.")

# -------------------------------------------
# 4. Ensure numeric columns have numeric dtype
# -------------------------------------------
for col in X_train_clean.columns:
    if col not in cat_cols:
        X_train_clean[col] = pd.to_numeric(X_train_clean[col], errors="coerce")
        X_test_clean[col] = pd.to_numeric(X_test_clean[col], errors="coerce")

print("✅ Numeric columns coerced to float.")
```

```python
del(X_train)
del(X_test)
```

```python
from catboost import CatBoostRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
import pandas as pd

# -------------------------------
# 1. Identify categorical columns
# -------------------------------
cat_cols = X_train_clean.select_dtypes(include=['string', 'object']).columns.tolist()
cat_idx  = [X_train_clean.columns.get_loc(c) for c in cat_cols]

print("Categorical:", cat_cols)


model = CatBoostRegressor(
    iterations=500,
    learning_rate=0.05,
    depth=8,
    loss_function='RMSE',
    eval_metric='RMSE',
    random_seed=42,
    verbose=100,
    early_stopping_rounds=50
)

model.fit(
    X_train_clean, y_train,
    cat_features=cat_idx,
    eval_set=(X_test_clean, y_test),
    use_best_model=True
)

# -------------------------------
# 4. Evaluation
# -------------------------------
y_pred = model.predict(X_test_clean)

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae  = mean_absolute_error(y_test, y_pred)
r2   = r2_score(y_test, y_pred)

print(f"RMSE: {rmse:.2f}")
print(f"MAE : {mae:.2f}")
print(f"R²  : {r2:.3f}")
```

```python
import pandas as pd
```

```python
fl = pd.read_parquet("flights_2018_final_encoded.parquet")
```

```python
vars = pd.read_parquet("target_variables.parquet")
```

```python
fl.drop(columns=['Dep_DateTime', 'Arr_DateTime', 'FlightID_1', '__index_level_0__'], inplace=True)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

# ------------------------------------------------
# 1. Identify categorical columns (object/string)
# ------------------------------------------------
cat_cols = fl.select_dtypes(include=["object", "string"]).columns.tolist()
print("Categorical columns:", cat_cols)

# ------------------------------------------------
# 2. Clean categorical missing values (In-place)
# ------------------------------------------------
# We iterate directly on fl columns to avoid creating a new DataFrame
for col in cat_cols:
    fl[col] = fl[col].astype("string")
    fl[col] = fl[col].fillna("missing")
    fl[col] = fl[col].replace({"<NA>": "missing", "nan": "missing", "None": "missing"})

# ------------------------------------------------
# 3. Label encode categorical columns (In-place)
# ------------------------------------------------
encoders = {}

for col in cat_cols:
    le = LabelEncoder()

    # Fit on the current column in fl
    le.fit(fl[col])

    # Ensure "missing" is included in encoder classes for robustness
    if "missing" not in le.classes_:
        le.classes_ = np.append(le.classes_, "missing")

    # Transform fl column directly
    fl[col] = le.transform(fl[col])
    encoders[col] = le

print("✅ Label encoding complete for all categorical variables.")

# ------------------------------------------------
# 4. Convert all non-categorical columns to numeric (In-place)
# ------------------------------------------------
for col in fl.columns:
    if col not in cat_cols:
        fl[col] = pd.to_numeric(fl[col], errors="coerce")

print("✅ All numeric columns coerced to float.")
print("🎉 Final dataframe shape:", fl.shape)
```

```python
print(fl["target"].value_counts())
```

```python
from sklearn.utils import resample

# Get indices per class (cheap, only index arrays)
idx_0 = fl.index[fl["target"] == 0].to_numpy()
idx_1 = fl.index[fl["target"] == 1].to_numpy()
idx_2 = fl.index[fl["target"] == 2].to_numpy()
idx_3 = fl.index[fl["target"] == 3].to_numpy()
idx_4 = fl.index[fl["target"] == 4].to_numpy()

# Decide how many of class 0 to keep
# Here: match class 1 size (you can tweak this)
n_0_keep = len(idx_1)
rng = np.random.default_rng(42)
idx_0_down = rng.choice(idx_0, size=n_0_keep, replace=False)

# Other classes: keep all indices
final_idx = np.concatenate([idx_0_down, idx_1, idx_2, idx_3, idx_4])

# Build balanced dataframe (one big materialization step)
fl_balanced = fl.loc[final_idx].sample(frac=1.0, random_state=42).reset_index(drop=True)

# Free original big df
del fl

print("Balanced class counts:\n", fl_balanced["target"].value_counts())
print("Balanced shape:", fl_balanced.shape)
```

```python
X = fl_balanced.drop(columns=["target", "DelayBucket", "DepDelay"], errors="ignore")
y = fl_balanced["target"]
del(fl_balanced)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    random_state=420,
    stratify=y
)

print("X_train:", X_train.shape)
print("X_test :", X_test.shape)
print(y_train.value_counts(), "\n", y_test.value_counts())
```

```python
from catboost import CatBoostClassifier
from sklearn.metrics import accuracy_score, classification_report

model = CatBoostClassifier(
    iterations=10,
    learning_rate=0.1,
    depth=10,
    loss_function="MultiClass",
    eval_metric="TotalF1",
    random_seed=420,
    verbose=1,
    early_stopping_rounds=50
)

model.fit(
    X_train, y_train,
    eval_set=(X_test, y_test),
    use_best_model=True
)

y_pred = model.predict(X_test)

print("\nAccuracy:", accuracy_score(y_test, y_pred))
print("\nClassification report:\n")
print(classification_report(y_test, y_pred))
```

```python
class_weights = {
    0: 0.75,
    1: 1.0,
    2: 1.5,
    3: 2.5,
    4: 4.0
}

model = CatBoostClassifier(
    iterations=400,
    depth=10,
    learning_rate=0.15,
    loss_function="MultiClass",
    eval_metric="TotalF1",
    class_weights=class_weights,
    random_seed=42,
    verbose=1
)

model.fit(
    X_train, y_train,
    eval_set=(X_test, y_test),
    use_best_model=True
)
```
